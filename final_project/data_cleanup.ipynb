```python
# load libraries
import pandas as pd
import numpy as np
import glob
```

```python
# import .csv fiels from data folder on Google Drive
path = r'/content/drive/MyDrive/2021_Fall_USF_PMS_Biotechnology/MSDS_596/final_project/data/' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, error_bad_lines=False)
    li.append(df)

ks_df = pd.concat(li, axis=0, ignore_index=True)
ks_df.head(5)
```
